##### Observation Outcome:
 #####   1- the columns: job, education, contact and poutcome are the only ones that contain 'unknown' and should be replaced with NA
   ##### 2- column 'y' should be renamed to 'deposited'.

```python
bank_df.rename(columns = {'y':'deposited?'}, inplace = True)
```

```python
#replacing values
bank_df['job'].replace('unknown', np.nan, inplace = True)
```

```python
bank_df['education'].replace('unknown', np.nan, inplace = True)
```

```python
bank_df['contact'].replace('unknown', np.nan, inplace = True)
```

```python
bank_df['poutcome'].replace('unknown', np.nan, inplace = True)
```

After observing, more than half of the data set has been deleted due to missing info.
As a result, NaN in job and education will be replaced with other.
Replace Nan in contact with the mode value.

```python
bank_df.mode()
```

```python
bank_df['job'].replace(np.nan, 'other', inplace = True)
```

```python
bank_df['contact'].replace(np.nan, 'cellular', inplace = True)
```
